Lab 2.1: Number Classifier


```python
# === 1. รับข้อมูลจากผู้ใช้ และ แปลงเป็นจำนวนเต็ม ===
# (Responsible AI Note: ใช้ try-except เพื่อป้องกันโปรแกรมล่มหากผู้ใช้พิมพ์สิ่งที่ไม่ใช่ตัวเลข)
try:
    user_input = input("กรุณากรอกตัวเลขจำนวนเต็ม: ")
    number = int(user_input)

    # กำหนดตัวแปรสำหรับเก็บคำอธิบายผลลัพธ์ (Explainable AI)
    sign_result = ""
    parity_result = ""

    # === 2. ตรวจว่าเป็นบวก, ลบ, หรือ ศูนย์ ด้วย IF-ELIF-ELSE ===
    if number > 0:
        sign_result = "เป็นจำนวนบวก"
    elif number < 0:
        sign_result = "เป็นจำนวนลบ"
    else:
        sign_result = "เป็นศูนย์"

    # === 3. ตรวจว่าเป็นเลขคู่หรือเลขคี่ ===
    # เลขคู่คือเลขที่หารด้วย 2 แล้วเหลือเศษ 0 (ใช้เครื่องหมาย %)
    if number % 2 == 0:
        parity_result = "เลขคู่"
    else:
        parity_result = "เลขคี่"

    # === 4. (CHALLENGE) รวมผลทั้งสองเข้าด้วยกัน ===
    # รวมข้อความการตัดสินใจเพื่อให้ผู้ใช้งานอ่านและเข้าใจได้ง่ายในทันที
    if number == 0:
        # เลข 0 ถือว่าเป็นเลขคู่ แต่ในทางตรรกะเรามักบอกแค่ว่าเป็นศูนย์และเป็นเลขคู่
        final_output = f"ตัวเลข {number} คือ 'ศูนย์' และเป็น '{parity_result}'"
    else:
        final_output = f"ตัวเลข {number} {sign_result} และเป็น '{parity_result}'"

    # แสดงผลลัพธ์ที่ชัดเจนและอธิบายได้ (Explainable Output)
    print("\n--- ผลลัพธ์การตรวจสอบ ---")
    print(final_output)

except ValueError:
    # ดักจับกรณีผู้ใช้พิมพ์ข้อความที่ไม่ใช่ตัวเลข เช่น "hello" หรือ "12.5"
    print("\n ข้อผิดพลาด: กรุณากรอกเฉพาะตัวเลขจำนวนเต็มเท่านั้น!")
```

Lab 2.2: Nested Movie Recommendation Engine

```python
# === 1. รับข้อมูลอายุและแปลงเป็นจำนวนเต็ม ===
try:
    age_input = input("กรุณากรอกอายุของคุณ (ปี): ")
    age = int(age_input)

    # (Responsible AI) ตรวจสอบความสมเหตุสมผลของข้อมูลก่อนประมวลผล
    if age < 0 or age > 120:
        print(" ข้อผิดพลาด: อายุไม่อยู่ในเกณฑ์ที่สมจริง กรุณารันโปรแกรมใหม่")
    else:
        # === 2. (CHALLENGE) ถามความชอบเพื่อนำมาคิดร่วมกับอายุ ===
        # รับข้อมูลความชอบหนังแอ็กชัน (แปลงเป็นตัวพิมพ์เล็ก .lower() เพื่อให้โค้ดทำงานได้ง่ายขึ้น)
        like_action = input("คุณชอบดูหนังแอ็กชันไหม? (ใช่ / ไม่ใช่): ").strip()

        # สร้างตัวแปรเก็บผลลัพธ์คำแนะนำ (Explainable AI)
        recommended_movie = ""
        reason = ""

        # === 3. ตรวจสอบเงื่อนไขด้วย IF-ELIF-ELSE และ AND ===
        # สมมติตรรกะการแบ่งเรตภาพยนตร์ที่เป็นธรรมและเหมาะสมกับวัย (AI Ethics)

        if age >= 18 and like_action == "ใช่":
            recommended_movie = "Mad Max: Fury Road (เรต R)"
            reason = "คุณอายุ 18 ปีขึ้นไป และชอบความตื่นเต้นของหนังแอ็กชัน"

        elif age >= 18 and like_action != "ใช่":
            recommended_movie = "The Godfather (เรต R)"
            reason = "คุณอายุ 18 ปีขึ้นไป แต่ไม่เน้นสายแอ็กชันดุเดือด เราจึงแนะนำหนังดราม่าอาชญากรรมระดับคลาสสิก"

        elif age >= 13 and like_action == "ใช่":
            recommended_movie = "The Avengers (เรต PG-13)"
            reason = "คุณอายุเกิน 13 ปี เหมาะกับเนื้อหาแอ็กชันฮีโร่ที่สนุกสนานแต่ไม่รุนแรงเกินไป"

        elif age >= 13 and like_action != "ใช่":
            recommended_movie = "Harry Potter (เรต PG-13)"
            reason = "คุณอายุเกิน 13 ปี แต่ไม่ได้ชอบหนังแอ็กชันจ๋า เราจึงแนะนำหนังแนวแฟนตาซีผจญภัยแทน"

        else:
            # สำหรับเด็กอายุต่ำกว่า 13 ปี (AI Ethics: ปกป้องเด็ก ไม่แนะนำหนังรุนแรงเด็ดขาด ไม่ว่าจะชอบแอ็กชันหรือไม่)
            recommended_movie = "Finding Nemo (เรต G)"
            reason = "เนื่องจากคุณอายุต่ำกว่า 13 ปี เราจึงแนะนำภาพยนตร์แอนิเมชันที่ดูได้ทุกเพศทุกวัยเพื่อความปลอดภัย"

        # === 4. แสดงผลลัพธ์ที่อธิบายได้ (Explainable Output) ===
        print("\n--- ระบบแนะนำภาพยนตร์ ---")
        print(f" ภาพยนตร์ที่แนะนำ: {recommended_movie}")
        print(f" เหตุผล: {reason}")

except ValueError:
    print("\n ข้อผิดพลาด: กรุณากรอกอายุเป็นตัวเลขจำนวนเต็มเท่านั้น!")
```

ACTIVITY 2.1

```python
# === 1. รับข้อมูลรายได้และแปลงเป็นจำนวนเต็ม ===
try:
    income_input = input("กรุณากรอกรายได้ต่อปีของคุณ (บาท): ")
    income = int(income_input)

    # (Responsible AI) ตรวจสอบความสมเหตุสมผลของข้อมูลรายได้
    if income < 0:
        print(" ข้อผิดพลาด: รายได้ไม่สามารถติดลบได้ กรุณารันโปรแกรมใหม่")
    else:
        # กำหนดตัวแปรสำหรับเกณฑ์ภาษีและการอธิบาย (Explainable AI)
        tax_rate = 0
        tax_amount = 0

        # === 2. คำนวณอัตราภาษีด้วย IF-ELIF-ELSE ===
        # อ้างอิงตามเกณฑ์ขั้นบันไดที่เป็นธรรมในสังคม (AI Ethics)
        if income <= 150000:
            tax_rate = 0
        elif income <= 300000:
            tax_rate = 5
        elif income <= 500000:
            tax_rate = 10
        else:
            tax_rate = 15

        #คำนวณมูลค่าภาษีที่ต้องจ่าย
        tax_amount = (income * tax_rate) / 100

        # === 3. (CHALLENGE) ใช้ Nested IF เพื่อตรวจสอบสถานะและให้ส่วนลดภาษี ===
        # สมมติเงื่อนไข: หากเป็น "นักศึกษา" จะได้รับส่วนลดภาษีเพิ่มเติมเพื่อสนับสนุนการศึกษา
        is_student = input("คุณเป็นนักศึกษาอยู่หรือไม่? (ใช่ / ไม่ใช่): ").strip()

        discount = 0
        if is_student == "ใช่":
            # ถ้ามีภาษีต้องจ่าย จะให้ส่วนลด 10% ของยอดภาษีนั้น
            if tax_amount > 0:
                discount = tax_amount * 0.10
                tax_amount = tax_amount - discount

        # === 4. แสดงผลลัพธ์ที่ชัดเจนและอธิบายได้ (Explainable Output) ===
        print("\n---  ใบสรุปการคำนวณภาษีประจำปี ---")
        print(f" รายได้รวมต่อปี: {income:,} บาท")
        print(f" อัตราภาษีของคุณ: {tax_rate}%")

        if discount > 0:
            print(f" ส่วนลดพิเศษ (ฐานะนักศึกษา 10%): -{discount:,} บาท")

        print(f" จำนวนภาษีสุทธิที่ต้องชำระ: {tax_amount:,} บาท")

        # อธิบายที่มาที่ไปให้ผู้ใช้เข้าใจแจ่มแจ้ง
        print("\n หมายเหตุการประมวลผล:")
        if tax_rate == 0:
            print("  -> คุณได้รับการยกเว้นภาษีเนื่องจากรายได้ไม่เกิน 150,000 บาท")
        else:
            print(f"  -> ภาษีคำนวณจากฐานรายได้ {tax_rate}% ลักลั่นตามเกณฑ์มาตรฐาน")

except ValueError:
    print("\n ข้อผิดพลาด: กรุณากรอกรายได้เป็นตัวเลขจำนวนเต็มเท่านั้น!")
```

```python
# === 1. รับข้อมูลจากผู้ใช้จำนวน 2 ตัวแปร และแปลงเป็นจำนวนเต็ม ===
try:
    var1_input = input("กรุณากรอกตัวเลขสำหรับตัวแปรที่ 1 (A): ")
    var2_input = input("กรุณากรอกตัวเลขสำหรับตัวแปรที่ 2 (B): ")

    A = int(var1_input)
    B = int(var2_input)

    # บันทึกค่าเริ่มต้นไว้เพื่อใช้ในการอธิบายผลลัพธ์ภายหลัง (Explainable AI)
    original_A = A
    original_B = B

    swapped = False # ตัวแปรช่วยจำสถานะเพื่อความโปร่งใส

    # === 2. ตรวจสอบเงื่อนไขเพื่อสลับค่าด้วย IF-ELSE ===
    # โจทย์: ถ้า A มากกว่า B ให้ทำการสลับค่ากัน เพื่อให้ A มีค่าน้อยกว่า B เสมอ
    if A > B:
        # ใช้แนวทางที่ปลอดภัย (Responsible): ใช้ตัวแปรสำรอง (Temp) เพื่อไม่ให้ข้อมูลสูญหาย
        temp = A
        A = B
        B = temp
        swapped = True
    else:
        # ถ้า B มากกว่าหรือเท่ากับ A อยู่แล้ว ไม่ต้องสลับค่า
        swapped = False

    # === 3. (CHALLENGE) ใช้ IF อีกชุดเพื่อแจ้งสถานะว่ามีการสลับค่าเกิดขึ้นหรือไม่ ===
    print("\n--- 🔄 ระบบตรวจสอบและปรับสมดุลข้อมูล ---")

    if swapped:
        print(f" สถานะ: เกิดการสลับค่าข้อมูล! (เนื่องจากค่าเริ่มต้น A มากกว่า B)")
        print(f" อธิบายการเปลี่ยนแปลง:")
        print(f"   - ค่าเดิมก่อนประมวลผล: A = {original_A}, B = {original_B}")
        print(f"   - ค่าใหม่หลังปรับสมดุล: A = {A}, B = {B}")
    else:
        print(f" สถานะ: ไม่มีการสลับค่าข้อมูล (เนื่องจากค่าเริ่มต้นเรียงลำดับจากน้อยไปมากถูกต้องอยู่แล้ว)")
        print(f" อธิบายสถานะ:")
        print(f"   - ค่าปัจจุบัน: A = {A}, B = {B} (ตรงตามเกณฑ์ที่ A ต้องน้อยกว่าหรือเท่ากับ B)")

except ValueError:
    print("\n ข้อผิดพลาด: กรุณากรอกเฉพาะตัวเลขจำนวนเต็มทั้งสองค่าเท่านั้น!")
```
